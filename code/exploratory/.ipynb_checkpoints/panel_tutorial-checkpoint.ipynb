```python
import panel as pn 
import numpy as np 
import pandas as pd 
import altair as alt 
import gcutils
colors, palette = gcutils.viz.altair_style()
pn.extension('vega')
```

```python
def get_data(amplitude, frequency):
    x = np.linspace(-2 * np.pi, 2 * np.pi, 300)
    y = amplitude * np.sin(frequency * x)
    df = pd.DataFrame([])
    df['x'] = x
    df['y'] = y
    return df

def get_plot(amplitude, frequency):
    df = get_data(amplitude, frequency)
    base = alt.Chart(df).encode(
            x=alt.X(field='x', type='quantitative', title='x axis'),
            y = alt.Y(field='y', type='quantitative', title='y axis'))
    line = base.mark_point().interactive()
    return line
```

```python
interact = pn.interact(get_plot, amplitude)
```

```python
out = get_plot(1, 1)
out
```
